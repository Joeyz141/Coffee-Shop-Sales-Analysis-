# ☕ Coffee Shop Sales Analysis Dashboard

## Project Overview

In this project we explore coffee shop sales in three locations to understand customer relations with coffee shop products. We look at the revenue generated and transactions purchased over time to better understand customer demands. In this process, we use Excel to prepare the data for analysis, then explore the data with pivot tables, and lastly build a dynamic dashboard. We then finalize our dashboard in PowerBI and add additional metrics and filters so that stakeholders can interact and analyze KPIs, sales trends, and product performance.

Overall, for coffee shop stakeholders to make beneficial business decisions, we use data-driven insights to help analyze and understand the business to increase its revenue through its strengths and weaknesses showcased via data in our interactive dashboard.

---

# Business Problem

Coffee shops operate in a highly competitive environment where businesses face high operating expenses, thin profit margins, and operational inefficiencies. Without data-driven insights into their sales performance, it becomes difficult for stakeholders to identify demand patterns, allocate resources efficiently, and make business decisions driven by data.
Coffee shops need better visibility into sales performance across locations, products, and time periods via data to make the right business decisions to improve overall business performance.

# Data Preparation in Excel

Utilized feature engineering to create new fields that give us columns **Revenue** (price * quantity), columns **Month**, **Day of the week**, and **Hour** extracted from transaction_date.

---

# Exploratory Analysis Using Pivot Tables

Pivot table analysis was used to identify key sales patterns across time, product categories, and transaction volume. This revealed strong seasonal trends, thus highlighting which products drive the majority of transactions.


Key findings from the exploratory analysis include:
- **Seasonal revenue trends:** sales are slower during winter periods, January & February, but sales increase steadily from March through June, indicating stronger customer demand during spring/summer seasons.
- **Peak demand hours:** transaction volumes are highest on weekdays from 7AM to 10AM, while activity declines during evening hours from 6PM until closing, suggesting business is driven by morning commuter demand.
- **Product category performance:** Coffee and Tea dominate product sales, while categories such as Packaged Chocolate, Branded Products, Loose Tea, and Coffee Bean account for less than 5% of total items sold, indicating lower demand for these product lines.

# Key Insights

### Weekday Demand is Higher

- Sales activity is consistently higher on weekdays compared to weekends, suggesting that commuter traffic is a core driver of daily customer demand. 

### Astoria Store Performs Efficiently

- Astoria opens from 7AM to 7PM, but other stores operate from 6AM to 8PM. However, Astoria still generates the second highest revenue, meaning Astoria operates more efficiently with its sales.

### Winter Season Sales Dip

- Monthly revenue showcases January & February having a noticeable dip in sales, followed by growth in sales from March to June.

### Lower Manhattan Evening Demand is Low

- After **7pm**, the Lower Manhattan location experiences minimal sales, indicating there is a low evening demand compared to other stores.


### Morning Rush Drives Demand

- There is high demand from the hours of 7AM to 10AM, suggesting strong morning commuter traffic.

---

# Business Recommendations

### 1. Adjust Lower Manhattan Store Hours

- Closing the Manhattan store at 7pm could reduce operational costs without revenue being affected.

### 2. Expand Morning Staffing

- Expanding morning staffing from 7AM to 10AM would improve service speed and reduce wait times, increasing customer satisfaction in meeting strong commuter traffic demand.

### 3. Introduce Winter Promotions

- Introduce winter promotions to stimulate more traffic during slow winter periods.

### 4. Increase Budgets for Coffee & Tea

- Since coffee and tea account for over **70%** of total items sold, increase marketing budgets, expand menu offerings, and prioritize inventory for these categories while reducing investment in lower-performing products to maximize revenue growth.
