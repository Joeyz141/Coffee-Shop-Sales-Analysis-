# ☕ Coffee Shop Sales Analysis Dashboard

## Project Overview

This project analyzes transactional sales data from a multi-location coffee shop using **Microsoft Excel and Power BI**. The goal was to transform raw transactional data into actionable business insights through a structured analytics workflow.

The analysis began in **Excel**, where the dataset was cleaned, transformed, and explored using pivot tables and charts. After understanding the core trends in the data, the analysis was extended in **Power BI** to build an interactive dashboard designed for stakeholders to monitor business performance and explore sales insights dynamically.

The project demonstrates a **complete analytics workflow**:

1. Data preparation and feature creation  
2. Exploratory data analysis using pivot tables  
3. Visualization and trend analysis  
4. Interactive dashboard development  

---

# Business Problem

Coffee shop management needs better visibility into sales performance across locations, products, and time periods. Without a clear analytical view of transactional data, it is difficult to understand customer demand patterns, optimize staffing schedules, and identify which products drive the most revenue.

---

# Project Objective

The objective of this project was to analyze coffee shop transaction data and develop an interactive dashboard that helps stakeholders:

- Monitor key business performance metrics  
- Identify peak demand periods  
- Evaluate product performance  
- Support operational and marketing decisions with data-driven insights

# Data Preparation in Excel

Before building the Power BI dashboard, the dataset was first explored and prepared in Excel.

Key preparation steps included:

- Creating a **Revenue column** by multiplying product price and transaction quantity
- Extracting **Month**, **Day of Week**, and **Hour of Day** from transaction timestamps
- Formatting data types for accurate analysis
- Applying filters to explore transaction patterns across locations and products

These transformations allowed the dataset to be analyzed across **time-based and product-based dimensions**.

---

# Exploratory Analysis Using Pivot Tables

Pivot tables were used to quickly summarize and understand the dataset before building the final dashboard.

Exploratory analyses included:

- **Revenue by Month** to identify seasonal trends
- **Transactions by Day of Week** to evaluate weekday vs weekend demand
- **Transactions by Hour of Day** to identify peak sales periods
- **Transactions by Product Category** to understand product mix
- **Top 15 Products** to evaluate which items generated the most transactions and revenue

These pivot tables helped identify key trends that informed the final Power BI dashboard design.

---

# Dynamic Dashboard in Excel

# Key Insights

### Weekday Demand is Higher

Sales activity is consistently higher on weekdays compared to weekends, indicating strong commuter-driven demand.

### Astoria Store Performs Efficiently

The Astoria location operates fewer hours (**7 AM–7 PM**) compared to other stores (**6 AM–8 PM**) yet still produces the **second highest revenue**, demonstrating strong sales efficiency during its operating hours.

### February Sales Dip

Monthly revenue trends show a noticeable dip in February followed by steady growth toward the spring and summer months.

### Lower Manhattan Evening Demand is Low

The Lower Manhattan location experiences minimal sales after **7 PM**, indicating low evening demand compared to other stores.

### Coffee and Tea Drive Sales

Coffee and Tea products represent the largest share of transactions across all store locations.

### Morning Rush Drives Demand

Transaction volume peaks during **7 AM–10 AM**, suggesting strong morning commuter traffic.

---

# Business Recommendations

### 1. Adjust Lower Manhattan Store Hours

Closing the Lower Manhattan location at **7 PM** could reduce operational costs without significantly affecting revenue.

### 2. Increase Morning Staffing

Since peak demand occurs between **7 AM and 10 AM**, increasing staff during these hours could improve service speed and customer satisfaction.

### 3. Introduce Winter Promotions

Sales decline during January and February compared to later months. Seasonal promotions during winter could help stimulate demand during slower periods.

### 4. Prioritize Core Categories

Coffee and Tea should remain central to marketing campaigns, product placement, and inventory planning since they represent the strongest-performing product categories.
