# ☕ Coffee Shop Sales Analysis Dashboard

## Project Overview

In this project we explore coffee shop sales in three locations to understand customer relations with coffee shop products. We look at the revenue generated and transactions purchased over time to better understand customer demands. In this process we use Excel to prepare the data for analysis, then explore the data with pivot tables, and lastly build a dynamic dashboard. We then finalize our dashboard in PowerBI and add additional metrics and filters so that stakeholders can interact and analyze KPIs, sales trends, and product performance.

Overall, for coffee shop stakeholders to make beneficial business decisions, we use data-driven insights to help analyze and understand the business to increase its revenue through its strengths and weaknesses showcased via data in our interactive dashboard.

---

# Business Problem

Coffee shops need better visibility into sales performance across locations, products, and time periods. These factors are needed in making the right business decisions for stakeholders.

# Data Preparation in Excel

## Objective 1: Prepare the Data for Analysis


- Familiarize yourself with the data
- Add a new column to caculate **Revenue** (price * quantity)
- Add new columns to calculate the **Month** and **Day of the week** based on transaction_date
- Add a new column to extract the hour from the transaction_date

These new datasets now allows for our data to be analyzed across time and its performance via revenue made and transactions purchased.

---

# Exploratory Analysis Using Pivot Tables

- Insert a PivotTable on a new sheet to show **revenue** by **month**
- Add two more PivotTables showing **number of transactions** by **day of week** and **hour of day**
- Add a PivotTable to show the **number of transactions** by **product category**, sorted descending by transactions
- Add a PivotTable to show the **number of transactions** and **revenue** by **product type**, sorted descending and filtered to the Top 15


These pivot tables allow us to see our sales performance and get us ready in order to make an interactive dashboard via Excel.

---

# Dynamic Dashboard in Excel

- Transform revenue by month as a line chart, transactions by day of week and hour of day as column charts, and transactions by product category as a bar chart
- Assemble the charts into a rough dashboard layout, including PivotTable
- Add a **slicer** for the store location, and report it to all pivot tables.


These interactive graphs allowed us to now see a visualization of our sales performance across our three coffee shop locations.

# Key Insights

### Weekday Demand is Higher
- Sales activity is consistently higher on weekdays compared to weekends, indicating strong demand from commuters.

### Astoria Store Performs Efficiently

- Astoria opens from **7AM to 7PM,** but other stores operate from 6AM to 8PM. However, Astoria still generates the **second highest revenue**, meaning Astoria operates more **efficiently** with its sales.

### February Sales Dip

- Monthly revenue showcases February having a noticeable dip in sales, followed by growth in sales from March to June.

### Lower Manhattan Evening Demand is Low

- After 7pm, the lower manhattan location experiences minimal sales indicating there is a low evening demand compared to other stores.

### Coffee and Tea Drive Sales

- **Coffee** and **Tea** are the largest shares of transactions across all stores.

### Morning Rush Drives Demand
- There is high demand from the hours of **7AM to 10AM**, suggesting strong morning commuter traffic.

---

# Business Recommendations

### 1. Adjust Lower Manhattan Store Hours

- Closing the Manhattan store at **7pm** could reduce operational costs without revenue being affected

### 2. Increase Morning Staffing

- An increase in morning staff could increase sales in meeting strong commuter traffic 

### 3. Introduce Winter Promotions

- Introduce winter promotions to stimulate demand during slow winter periods

### 4. Prioritize Core Categories

- Coffee and Tea should remain central to marketing campaigns and stores' inventory since they are the store's strongest-performing product categories
